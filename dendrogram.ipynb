```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_data= pd.read_csv(r'C:\data set\Mall_Customers.csv')
customer_data.head()
```

```python
data = customer_data.iloc[:,3:5].values
data
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,8))
plt.title('cutomer data')
dnd = shc.dendrogram(shc.linkage(data,method='ward'))
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,8))
plt.title('cutomer data')
dnd = shc.dendrogram(shc.linkage(data,method='complete'))
```
